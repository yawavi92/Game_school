```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# groupby gender
gender_group = purchase_data.groupby ("Gender")["SN"].unique()
players_count = len(purchase_data["SN"].unique())
#find the total of male players
male_count =len( gender_group["Male"])

# # # # # #Perncetage of male players
per_male = (male_count/ players_count)*100

# # # # # #find total of female
female_count = len(gender_group["Female"])

# # # # # #Perncetage of female players
per_female = (female_count/players_count)*100

# # # # # #count for other/Non-Disclosed
other_count = len(gender_group["Other / Non-Disclosed"])

# # # # # #percentage other
per_other =  (other_count/players_count)*100

gender = pd.DataFrame({"":["Male", "Female","Other / Non-Disclosed"],
                             "Total Count":[male_count, female_count, other_count],
                              "Percentage of Players":[per_male, per_female, per_other]})
gender
# # # # # # Convert "gender_percent" column to percentage format
gender["Percentage of Players"] = gender["Percentage of Players"].map("{:,.2f}%".format)
# # # # # # # # Change the index to be the "Gender" column
gender.set_index("", inplace =True)

# # # # # # # # Display the DataFrame
gender
```

Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins for ages
bins = [0,9, 14, 19, 24, 29, 34, 39, max(purchase_data['Age'])]

#Categorize the existing players using the age bins. Hint: use pd.cut()
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Calculate the numbers by age group
purchase_data["Age_group"] = pd.cut(purchase_data["Age"], bins, labels=group_names,include_lowest = True, right = True)
purchase_data.head()
purchase_data_df = purchase_data.groupby("Age_group")["SN"].nunique()
purchase_df = purchase_data_df.to_frame('Total Count').sort_index()
purchase_df ["Percentage of Players"] = purchase_df['Total Count']/len(purchase_data["SN"].unique()) * 100
purchase_df

purchase_df ["Percentage of Players"] = purchase_df ["Percentage of Players"].map("{:,.2f}%".format)   

purchase_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price columns
Retrieve = purchase_data.filter(['Item ID', 'Item Name','Price'])
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value
purchase_count = Retrieve.groupby('Item ID')['Item Name'].value_counts()

avg = Retrieve.groupby(['Item ID','Item Name'])['Price'].mean()
total_purchase = avg*purchase_count
total_purchase
#Create a summary data frame to hold the results
Summary_data = purchase_count.to_frame("Purchase count")
Summary_data ["Item Price"] = avg.map("${:.2f}".format) 
Summary_data ["Total Purchase Value"] = total_purchase

#Sort the purchase count column in descending order
Summary_data = Summary_data.sort_values(by ='Purchase count', ascending=False)
Summary_data ["Total Purchase Value"] = Summary_data ["Total Purchase Value"].map("${:.2f}".format)
Summary_data.head()
#Optional: give the displayed data cleaner formatting

#Display a preview of the summary data frame
```
